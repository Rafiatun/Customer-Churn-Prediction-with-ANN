```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
%matplotlib inline
```

```python
file = pd.read_csv("churn.csv")
```

```python
file.drop('customerID',axis='columns',inplace=True)
```

```python
file[file.TotalCharges!=' '].shape
```

```python
file1 =file[file.TotalCharges!=' ']
file1.shape
```

```python
tenure_yes= file1[file1['Churn']=='Yes'].tenure
tenure_no= file1[file1['Churn']=='No'].tenure
plt.hist([tenure_yes ,tenure_no ] , color=['powderblue','blue'],label=['Yes','No'])
plt.legend()
```

```python
MonthlyCharges_yes= file1[file1['Churn']=='Yes'].MonthlyCharges
MonthlyCharges_no= file1[file1['Churn']=='No'].MonthlyCharges
plt.hist([MonthlyCharges_yes ,MonthlyCharges_no ] , color=['powderblue','blue'],label=['Yes','No'])
plt.legend()
```

```python
file1.replace('No internet service','No',inplace=True)
file1.replace('No phone service','No',inplace=True)
yes_no=['Partner','Dependents','PhoneService','MultipleLines','OnlineSecurity','OnlineBackup','DeviceProtection',
        'TechSupport','StreamingTV','StreamingMovies']
for num in yes_no:
    file1[num].replace({'Yes': 1 ,'No':0 }, inplace=True)
file1['gender'].replace({'Female': 1 ,'Male':0 }, inplace=True)
for i in file1:
    if file1[i].dtypes== 'object':
        print(f'{i:} {file1[i].unique()}')
file1.TotalCharges=pd.to_numeric(file1.TotalCharges)
```

```python
file1
```

```python
yes_no_2=['PaperlessBilling','Churn']

for num in yes_no_2:
    file1[num].replace({'Yes': 1 ,'No':0 }, inplace=True)
```

```python
file1.head(2)
```

```python
file1=pd.get_dummies(data=file1,columns=['InternetService','Contract','PaymentMethod'])
```

```python
file1.head(2)
```

```python
col_to_scale=['tenure','MonthlyCharges','TotalCharges']
from sklearn.preprocessing import MinMaxScaler
scaler= MinMaxScaler()
file1[col_to_scale]=scaler.fit_transform(file1[col_to_scale])
x= file1.drop('Churn',axis ='columns')
y= file1['Churn']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=40)
```

```python
import tensorflow as tf
from tensorflow import keras
```

```python
model=keras.Sequential([
    keras.layers.Dense(20 ,input_shape=(26,),activation='relu'),
    keras.layers.Dense(10 ,activation='relu'),
    keras.layers.Dense(1,activation='sigmoid')
])

model.compile(
optimizer='adam',
loss='binary_crossentropy',
metrics=['accuracy'])

model.fit(x_train,y_train,epochs=5)
```

```python
model=keras.Sequential([
    keras.layers.Dense(15 ,input_shape=(26,),activation='relu'),
    keras.layers.Dense(15 ,activation='relu'),
    keras.layers.Dense(1,activation='sigmoid')
])

model.compile(
optimizer='adam',
loss='binary_crossentropy',
metrics=['accuracy'])

model.fit(x_train,y_train,epochs=100)
```

```python

model.evaluate(x_test, y_test)
```

```python
y_pred = model.predict(x_test)
```

```python
y_predict = []
for element in y_pred:
    if element > 0.5:
        y_predict.append(1)
    else:
        y_predict.append(0)
```

```python
from sklearn.metrics import confusion_matrix , classification_report
```

```python
print(classification_report(y_test,y_predict))
```

```python
import seaborn as sn
cm = tf.math.confusion_matrix(labels=y_test,predictions=y_predict)

plt.figure(figsize = (10,7))
sn.heatmap(cm, annot=True, fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
